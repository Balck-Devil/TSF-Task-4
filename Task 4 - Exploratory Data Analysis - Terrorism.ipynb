## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

## Reading Data and Dasplaying It

```python
df = pd.read_csv("D:/Work/The Spark Foundation/Project 4/globalterrorismdb_0718dist.csv",encoding="ISO-8859-1")
df.head()
```

```python
df.shape
```

## Checking are there any Null values or not

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Removing Null values

```python
df.dropna(axis=1,how="all")
```

## Renaming Columns

```python
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","provstate":"State","region_txt":"Region","attacktype1_txt":"AttackType","target1":"Target","nkill":"Killed","nwound":"Wounded","summary":"Summary","gname":"Group","targtype1_txt":"Target_type","weaptype1_txt":"Weapon_type","motive":"Motive",
},inplace=True)

df.columns
```

## Selecting important Columns

```python
df = df[["Year","Month","Day","Country","State","Region","city","latitude","longitude","AttackType","Killed","Wounded","Target","Summary","Group","Target_type","Weapon_type","Motive"]]
```

# Data Analysis

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(),annot=True)
```

### 1. Top 5 Effected Countries

```python
names = df["Country"].value_counts().index[:5]
values = df["Country"].value_counts()[:5].values

plt.figure(figsize=(8,6))
sns.barplot(names, values, palette="bright")
plt.xlabel("Country")
plt.ylabel("Count")
plt.title("Top 5 Country")
```

### 2. Terrorism Activity Yearwise

```python
plt.figure(figsize=(10,5))
sns.countplot("Year", data=df,order=df["Year"].value_counts().index, palette="bright")
plt.xticks(rotation=90)
plt.show()
```

### 3. Top 5 Effected States

```python
names = df["State"].value_counts().index[:5]
values = df["State"].value_counts()[:5].values

plt.figure(figsize=(8, 6))
sns.barplot(names, values, palette="bright")
plt.xlabel("State")
plt.ylabel("Count")
plt.title("Top 5 State")
```

### 4. Most Used Attack Type

```python
plt.figure(figsize=(10,5))
sns.countplot("AttackType", data=df,order=df["AttackType"].value_counts().index, palette="bright")
plt.xticks(rotation=90)
plt.show()
```

### 5. Most Frequent Target Type

```python
plt.figure(figsize=(10,5))
sns.countplot("Target_type", data=df, order=df["Target_type"].value_counts().index, palette="bright")
plt.xticks(rotation=90)
plt.show()
```

### Most Effected Region

```python
names = df["Region"].value_counts().index[:5]
values = df["Region"].value_counts()[:5].values

plt.figure(figsize=(8,6))
sns.barplot(names, values, palette="bright")
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Top 5 Region")
```

```python
print("Country With Most Attacks - ",df["Country"].value_counts().idxmax())
print("City With Most Attacks - ",df["city"].value_counts().index[1])
print("Region With Most Attacks - ",df["Region"].value_counts().idxmax())
print("Year With Most Attacks - ",df["Year"].value_counts().idxmax())
print("Month With Most Attacks - ",df["Month"].value_counts().idxmax())
print("Group With Most Attacks - ",df["Group"].value_counts().index[1])
print("Most Attacks - ",df["AttackType"].value_counts().idxmax())
```
